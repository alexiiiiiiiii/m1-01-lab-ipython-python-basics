```python
data = [
    {'user_ID' : 1, 'event_Type' : 'login', 'duration_Seconds' : 64},
    {'user_ID' : 2, 'event_Type' : 'signup', 'duration_Seconds' : 15},
    {'user_ID' : 3, 'event_Type' : 'login', 'duration_Seconds' : 128},
    {'user_ID' : 4, 'event_Type' : 'login', 'duration_Seconds' : -20},
    {'user_ID' : 5, 'event_Type' : 'video_Play', 'duration_Seconds' : 'idk'},
    {'user_ID' : 6, 'event_Type' : 'login', 'duration_Seconds' : 684},
    {'user_ID' : 7, 'event_Type' : 'login', 'duration_Seconds' : -155},
    {'user_ID' : 8, 'event_Type' : 'login', 'duration_Seconds' : 54},
    {'user_ID' : 9, 'event_Type' : 'checkout', 'duration_Seconds' : 21},
    {'user_ID' : 10, 'event_Type' : 'login', 'duration_Seconds' : 47},
    {'user_ID' : 11, 'event_Type' : 'signout', 'duration_Seconds' : 78},
    {'user_ID' : 12, 'event_Type' : 'login', 'duration_Seconds' : 91}
]


cleaned_events = []

for event in data:
    duration = event.get('duration_Seconds')

    if isinstance(duration, int or float) and duration >= 0:
        cleanedEvent = event.copy()
        cleanedEvent['duration_minutes'] = duration / 60
        cleaned_events.append(cleanedEvent)

print(cleaned_events)
```
